```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# LIF 뉴런 파라미터 정의
t_max = 150e-3  # 초
dt = 1e-3       # 초
tau = 20e-3     # 초
el = -60e-3     # 미리볼트
vr = -70e-3     # 미리볼트
vth = -50e-3    # 미리볼트
r = 100e6       # 옴
i_mean = 25e-11 # 암페어
```

```python
# 막전위 시뮬레이션
step_end = 10   # 스텝 횟수
v = el          # 막전위

# 스텝 횟수 만큼 반복
for step in range(step_end):
    t = step * dt                               # 시간 계산
    i = i_mean * (1+np.sin((t*2*np.pi) / 0.01)) # 시냅스 입력전류 계산
    v = v + dt / tau * (el-v+r*i)               # 막전위 계산
    print(f'{t:.3f} {i:.4e} {v:.4e}')
```

```python
# 입력전류 그래프
step_end = 25

with plt.xkcd():                        # 과학만화 스타일 그래프
    plt.figure()                        # 그래프 생성
    plt.title("Synaptic Input $l(t)$")  # 타이틀 지정
    plt.xlabel("time (s)")              # x축 라벨 지정
    plt.ylabel("$l$ (A)")               # y축 라벨 지정

    for step in range(step_end):
        t = step * dt                               # 시간 t 계산
        i = i_mean * (1+np.sin((t*2*np.pi) / 0.01)) # 시냅스 입력전류 계산
        plt.plot(t, i, "ko")                        # 전류 i 그래프 그리기
    
    plt.show()
```

```python
# 막전위 그래프
step_end = int(t_max / dt)  # 스텝 횟수 초기화
v = el                      # v0 초기화, 누수전위(leaky potential)

with plt.xkcd():                            # 과학만화 스타일 그래프
    plt.figure()                            # 그래프 생성
    plt.title("$V_m$ with sinusoidal l(t)") # 그래프 제목 지정
    plt.xlabel("time (s)")                  # x축 라벨 지정
    plt.ylabel("$V_m$ (V)")                 # y축 라벨 지정

    for step in range(step_end):
        t = step * dt                                   # 시간 t 계산
        i = i_mean * (1 + np.sin((t*2*np.pi) / 0.01))   # 시냅스 입력전류 계산
        v = v + dt / tau * (el - v + r*i)               # 막전위 v 계산
        plt.plot(t, v, "k.")                            # 막전위 v 그래프 그리기
    
    plt.show()
```

```python
# 랜덤성이 추가된 막전위 그래프
# np.random.seed(2020)
step_end = int(t_max / dt)  # 스텝 횟수 초기화
v = el                      # v0 초기화, 누수전위(leaky potential)

with plt.xkcd():                            # 과학만화 스타일 그래프
    plt.figure()                            # 그래프 생성
    plt.title("$V_m$ with sinusoidal l(t)") # 그래프 제목 지정
    plt.xlabel("time (s)")                  # x축 라벨 지정
    plt.ylabel("$V_m$ (V)")                 # y축 라벨 지정

    for step in range(step_end):
        t = step * dt                                   # 시간 t 계산
        random_num = 2 * np.random.random() - 1         # -1 ~ +1 사이의 난수 생성
        i = i_mean * (1 + 0.1 * (t_max / dt)**(0.5)*random_num)   # 시냅스 입력전류 계산
        v = v + dt / tau * (el - v + r*i)               # 막전위 v 계산
        plt.plot(t, v, "k.")                            # 막전위 v 그래프 그리기
    
    plt.show()
```

```python
# 다중 수행 통계 (평균, 표준편차)
# np.random.seed(2020)
step_end = int(t_max / dt)  # 스텝 횟수
n = 50                      # 막전위 개수
v_n = [el] * n              # n개의 막전위 리스트로 초기화

with plt.xkcd():                            # 과학만화 스타일 그래프
    plt.figure()                            # 그래프 생성
    plt.title("$V_m$ with sinusoidal l(t)") # 그래프 제목 지정
    plt.xlabel("time (s)")                  # x축 라벨 지정
    plt.ylabel("$V_m$ (V)")                 # y축 라벨 지정

    for step in range(step_end):
        t = step * dt
        
        for j in range(0, n):
            i = i_mean * (1 + 0.1 * (t_max / dt)**(0.5) * (2* np.random.random() - 1))
            v_n[j] = v_n[j] + (dt / tau) * (el - v_n[j] + r*i)
        
        v_mean = sum(v_n) / n
        v_var_n = [(v - v_mean)**2 for v in v_n]
        v_var = sum(v_var_n) / (n-1)
        v_std = np.sqrt(v_var)

        plt.plot(n*[t], v_n, "k.", alpha=0.1)
        plt.plot(t, v_mean, "C0.", alpha=0.8, markersize=10)
        plt.plot(t, v_mean + v_std, "C7.", alpha=0.8)
        plt.plot(t, v_mean - v_std, "C7.", alpha=0.8)
    
    plt.show()
```

```python
# 막전위 계산
# np.random.seed(2020)

step_end = int(t_max / dt) - 1
t_range = np.linspace(0, t_max, num=step_end, endpoint=False)
v = el * np.ones(step_end)
i = i_mean * (1 + 0.1 * (t_max / dt) ** (0.5) * (2 * np.random.random(step_end) - 1))

for step, i_step in enumerate(i):
    if step == 0:
        continue

    v[step] = v[step - 1] + (dt / tau) * (el - v[step - 1] + r * i_step)

with plt.xkcd():                            # 과학만화 스타일 그래프
    plt.figure()                            # 그래프 생성
    plt.title("$V_m$ with random l(t)")     # 그래프 제목 지정
    plt.xlabel("time (s)")                  # x축 라벨 지정
    plt.ylabel("$V_m$ (V)")                 # y축 라벨 지정
    plt.plot(t_range, v, 'k')
    plt.show()
```

```python
# 넘파이 다중 수행 통계
# np.random.seed(2020)

step_end = int(t_max / dt)
t_range = np.linspace(0, t_max, num=step_end)
v_n = el * np.ones([n, step_end])
i = i_mean * (1 + 0.1 * (t_max / dt)**(0.5) * (2*np.random.random([n, step_end]) - 1))

for step in range(1, step_end):
    v_n[:, step] = v_n[:, step - 1] + (dt / tau) * (el - v_n[:, step-1] + r * i[:, step])

v_mean = np.mean(v_n, axis=0)
v_std = np.std(v_n, axis=0)

with plt.xkcd():                                # 과학만화 스타일 그래프
    plt.figure()                                # 그래프 생성
    plt.title("Multiple realizations of $V_m$") # 그래프 제목 지정
    plt.xlabel("time (s)")                      # x축 라벨 지정
    plt.ylabel("$V_m$ (V)")                     # y축 라벨 지정
    plt.plot(t_range, v_n.T, 'k', alpha=0.3)
    plt.plot(t_range, v_mean, 'C0', alpha=0.8, label='mean')
    plt.plot(t_range, v_mean + v_std, 'C7', alpha=0.8)
    plt.plot(t_range, v_mean - v_std, 'C7', alpha=0.8, label='mean $\pm$ std')
    plt.legend()
    plt.show()
```
